```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests
from bs4 import BeautifulSoup
```

```python
url = 'http://www2.2019seagames.com/countries/'
web = requests.get(url)
data = BeautifulSoup(web.content, 'html.parser')

daftar = []

for i in data.find_all('em'):
    daftar.append(i.text)
negara = daftar[0:-1:6]

gold = daftar[2:-1:6]
gold = [i.replace('Gold: ', '') for i in gold]
gold = list(map(int, gold))



url2 = 'https://rs.2019seagames.com/RS2019/mobiapp/MedalTally'
web2 = requests.get(url2)
data2 = BeautifulSoup(web2.content, 'html.parser')

daftar2 = []

for i in data2.find_all('small'):
    daftar2.append(i.text)
negara2 = daftar2[5:-1:5]

gold2 = daftar2[6:-1:5]
gold2 = [i.replace('Gold: ', '') for i in gold2]
gold2 = list(map(int, gold2))
```

```python
plt.figure(figsize=(20,10))
plt.style.use('seaborn')
plt.plot(negara[0:11], gold[0:11], 'o-', label='Gold 2017', color='r')
plt.scatter(negara[4], gold[4], marker='*', s=400, zorder=3, color='y', label='Champion')
plt.title('GOLD MEDAL TALLY SEA GAMES 2017')
plt.grid(True)
plt.xticks(rotation= 30)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.style.use('seaborn')
plt.plot(negara2, gold2, 'o-', label='Gold 2019', color='g')
plt.scatter(negara2[0], gold2[0], marker='*', s=400, zorder=3, color='y', label='Champion')
plt.title('GOLD MEDAL TALLY SEA GAMES 2019')
plt.grid(True)
plt.xticks(rotation= 30)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(20,10))

plt.subplot(121)
x,y,z = plt.pie(gold[0:11], labels= negara[0:11],
       startangle = 90,
       autopct = '%1.1f%%')
for i in z:
    i.set_color('w')
plt.title('GOLD% SEA GAMES 2017', fontweight="bold", fontsize=20)

plt.subplot(122)
x,y,z = plt.pie(gold2, labels= negara2,
       startangle = 90,
       autopct = '%1.1f%%')
for i in z:
    i.set_color('w')
plt.title('GOLD% SEA GAMES 2019', fontweight="bold", fontsize=20)

plt.show()
```
